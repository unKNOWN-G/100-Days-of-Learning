```python
# Computation Libraries
import numpy as np
import pandas as pd

# Stats Libraries
import statistics as stats

# Visualisazition Libraries
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# Preprocessing
from sklearn.preprocessing import LabelEncoder

# Modelling Libraries
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn import svm

# Evaluation Metrics Libraries
from sklearn.metrics import accuracy_score, f1_score, jaccard_score
from sklearn.metrics import classification_report, confusion_matrix
```

```python
df = pd.read_csv("./Dataset/Classification/svm_voice.csv")
print("Number of rows : {0}\nNumber of columns : {1}\nColumns : {2}".format(len(df),len(df.columns),df.columns))
df.head(5)
```

```python
# Label Encoder
label_encoder = LabelEncoder()
df['label'] = label_encoder.fit_transform(df['label'])
df['label'].value_counts()
```

```python
# Data Preprocessing
X = df.drop('label',axis=1)
Y = df['label']
train_x,test_x,train_y,test_y = train_test_split(X,Y,test_size = 0.33)
```

```python
model = svm.SVC(kernel = 'poly')
model.fit(train_x,train_y)
```

```python
pred = model.predict(test_x)
pred
```

```python
from sklearn import metrics
print("Train set Accuracy: ", accuracy_score(train_y, model.predict(train_x)))
print("Test set Accuracy: ", accuracy_score(test_y, pred))
```

```python
print(classification_report(test_y,pred))
```
